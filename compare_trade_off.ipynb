### Generate Trade-Off Curve between Timber Supply and Net System Carbon Emission

```python
import glob

# Path to the folder containing the CSV files
file_path = 'results/trade_off/curves/*.csv'

# Load all CSV files in the folder
csv_files = glob.glob(file_path)

# Plotting each file's data
plt.figure(figsize=(12, 6))

for file in csv_files:
    # Read the CSV file
    df = pd.read_csv(file)
    
    # Extract columns for plotting
    harvest_volume = df['harvest volume']
    net_emission = df['net emission']
    
    # Get the file name without path and ".csv" extension
    curve_name = "Minimize Emission " +file.split("/")[-1].replace(".csv", "")
    
    # Plot the trade-off curve
    plt.plot(net_emission, harvest_volume, marker='o', label=curve_name)

# Plot the stock curve on the same graph
plt.plot(e_s_30, h_s_30, marker='x', linestyle='--', color='red', label='Maximize Stock 0.0_30')
plt.plot(e_s_60, h_s_60, marker='x', linestyle='--', color='blue', label='Maximize Stock 0.0_60')
plt.plot(e_s_90, h_s_90, marker='x', linestyle='--', color='green', label='Maximize Stock 0.0_90')

# Adding labels and title
plt.xlabel('CO2 (ton)')
plt.ylabel('Volume (m)')
plt.title('Trade-Off Curves: Wood Supply vs. Forest Carbon')
plt.legend(title="Curves", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

# Display the plots
plt.show()
```
